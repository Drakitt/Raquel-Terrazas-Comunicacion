```python
from scipy.io.wavfile import write
```

```python
#Parámetros de la señal
```

```python
mps= 44100   #muestras por segundo
frec_portadora = 80  #hz
frec_moduladora = 0.89 #hz
duracion = 2.5  #s
```

```python
t1= np.arange(0,10, 0.01)   #Para la señal portadora
```

```python
t2= np.arange(0,10, 0.000001)   #Para la señal moduladora
```

```python

t3= np.arange(0.0, 44100, 0.1)  #Para la señal resultante
#t3= np.radians(t3)
```

```python
#Calculamos la señal sinusoidal de la portadora
```

```python

muestras = np.arange(mps * duracion)
```

```python
portadora = np.sin(2 * np.pi * frec_portadora * muestras / mps)
#portadora *= 0.3  #atenuación
portadora_ints = np.int16(portadora * 32767)
```

```python
#Calculamos la señal sinusoidal de la modulante y la señal modulada AM
```

```python
moduladora = np.sin(2 * np.pi * frec_moduladora* muestras / mps)
```

```python
señal_modulada_FM = np.sin(2*np.pi*(frec_portadora + 15*moduladora)*muestras/mps)
```

```python
#señal_modulada_FM *= 0.3
```

```python
señal_modulada_FM_ints = np.int16(señal_modulada_FM * 32767)
```

```python
# Generamos la señal modulada
```

```python
write('señal_modulada.wav', mps, señal_modulada_FM_ints)
```
